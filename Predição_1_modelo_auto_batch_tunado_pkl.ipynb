### Predição_1_modelo_auto_batch_tunado_pkl

Aqui vamos usar o modelo serializado em "Automóveis_v1", para realizar previsões a partir de uma base de dados, o que é conhecido com batch, e modelo tunado no lâmbda e no intercepto.

```python
import pandas as pd
model = pd.read_pickle("modelo_grid_auto.pkl") #importação do modelo para predição
df_predict = pd.read_excel("auto.xlsx") #importação da base para predição
```

```python
#precisamos garantir que as variáveis categóricas e numéricas, estejam de fato nesses formatos

df_predict[model["num_features"]] = df_predict[model["num_features"]].astype(float)
df_predict[model["cat_features"]] = df_predict[model["cat_features"]].astype(str)

df_predict = df_predict[model["uni_features"]].copy()
```

```python
#rodando os modelos de predição
model["model2"].predict(df_predict)

#guardando os valores na base

df_predict["predict"] = model["model2"].predict(df_predict)
df_predict
```
